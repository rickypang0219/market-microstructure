# Random Walk Model and Bachelier’s First Law

## Gaussian Random Walk 
The Gaussian Random walk is a simple model in statistics where at each time, the random walker either move $+1$ or $-1$ with equal probability $P = 0.5$. We can discribe a random walk model by a sum of random variable $X$ where 

$$
X = \begin{cases} +1 , P(+1) = 0.5 & \\ -1, P(-1) = 0.5 \end{cases}
$$
Suppose the random walker starts in the origin, we are interested in the position of a random walker after $N$ steps. Where it is given by 

$$
W = \sum_{i=1}^N X_{i} = X_1 + X_2 + \cdots + X_N
$$

```python
def sample_discrete_x():
    random_number = random.uniform(0, 1)
    if random_number <= 0.5:
        return 1.0
    else:
        return -1.0

def generate_w(N:int):
    return np.cumsum([sample_discrete_x() for _ in range(N)])
```

```python
W: list[float] = [ ]
N: int = 100
for i in range(N): 
    W.append(sample_discrete_x())
```

```python
plt.figure(figsize=(10,5))
plt.title(f"10 Random Walk Simulation")
for i in range(15):
    plt.plot([*range(N)], generate_w(N))
plt.xlabel('Number of Steps N')
```

# Random Walker Return and Gaussian Distribution

Recall the defintion of Random Walk: 

$$
W = \sum^{N}_{i=1} X_i  = X_1 + X_2 + \cdots + X_N
$$
By Central Limit Theorem, when $N \to \infty $, the probability distribution of random variable $W$ is Gaussian meaning that $W \sim \mathcal{N}(0, N)$ with mean $\mu = 0$ and standard deviation $\sigma = \sqrt{N}$


```python
def cal_return(N:int):
    w = np.array([sample_discrete_x() for _ in range(N)])
    return np.array([ (w[i]- w[i-1])  for i in range(1,N) ])
```

# Bachelier’s First Law 

Reference:
- http://euler.nmt.edu/~olegm/586/HYD7-16.pdf

The **variogram** of a random walk 

$$
\mathcal{V}(\tau) = \mathbb{E} [(p_{t + \tau} - p_t)^2]
$$

Suppose $P_t = W_t = \sum^{t}_{i=1} X_{i}$ 

$$
\begin{align}
\mathcal{V}(\tau) &= \mathbb{E} [(P_{t + \tau} - P_t)^2] \\ 
&=  \mathbb{E} [ (W_{t + \tau}  - W_{t})^2 ] \\ 
&=  \mathbb{E} \Big [ (\sum^{t + \tau}_{i = t + 1} X_i)^2  \Big]  \\ 
&= \sum^{t + \tau}_{i = t + 1} \mathbb{E} [ X_{i}^2 ] \\
&= \sum^{\tau}_{i = 1} \mathbb{E} [ X_{i}^2 ] \\
&= \tau \sigma^2
\end{align}
$$

where in the fourth line we use the IID assumption of $X_i$

```python
def calculate_difference(arr, N):
    diff = arr[N::N] - arr[:-N:N]
    return diff
    

def calculate_variogram(random_walk):

    # Initialize variables for lag distances and variogram values
    lag_distances = []
    variogram_values = []

    # Group squared differences based on different lag distances
    for lag in range(1, 200):
        differences = calculate_difference(random_walk, lag)
        squared_diffs = differences ** 2
        lag_distances.append(lag)
        variogram_value = np.mean(squared_diffs)
        variogram_values.append(variogram_value)

    return lag_distances, variogram_values
```

```python
lag_distances, variogram_values = calculate_variogram(generate_w(100000))
```

```python
plt.plot(lag_distances, variogram_values)
```

```python
def variogram(tau):
    return np.array([i * np.mean(sample_discrete_x()**2) for i in tau ])
    
tau = np.arange(1,100)

plt.plot(tau, variogram(tau))
```
